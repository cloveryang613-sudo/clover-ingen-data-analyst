## Basic pandas DataFrame Test

The following test confirms that pandas can create and display a small DataFrame.

```python
platform_test = pd.DataFrame({
    "platform": [
        "Fari",
        "Senpai",
        "Sentinel Prime AI",
        "Aido Rover",
        "Aido Humanoid"
    ],
    "primary_data_type": [
        "Interaction sessions",
        "Learning progress",
        "Security events",
        "Sensor telemetry",
        "Joint motion data"
    ]
})

platform_test
```

```python
battery_levels = np.array([100, 92, 85, 77, 68])

print("Battery levels:", battery_levels)
print("Average battery level:", battery_levels.mean())
```

```python
plt.figure(figsize=(8,4))
plt.plot(
    ["09:00", "10:00", "11:00", "12:00", "13:00"],
    battery_levels,
    marker="o"
)

plt.title("Sample Aido Rover Battery Trend")
plt.xlabel("Time")
plt.ylabel("Battery SoC (%)")
plt.show()
```

**Finding:** The sample battery level decreases over time, confirming that Matplotlib is working correctly.
